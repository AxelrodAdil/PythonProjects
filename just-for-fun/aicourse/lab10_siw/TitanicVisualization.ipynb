```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

df = pd.read_csv('titanic.csv')
df.describe()
```

```python
sns.countplot(x=df['survived'])
```

```python
sns.countplot(x=df['survived'], hue=df['sex'])
```

```python
sns.barplot(x='sex',y='age', data=df, hue='survived')
```

```python
sns.displot(data=df,x='survived',kind='hist',bins=2)
```

```python
sns.violinplot(x='pclass', y='age', hue='survived', split=True, data=df, palette={0: "r", 1: "g"});
```

```python
sns.barplot(x="embarked", y="fare", data=df)
plt.title("Mean ticket fare by embarking point \n")
plt.ylabel('Mean ticket fare')
plt.xlabel('Port')
```

```python
plt.figure(figsize=(12, 7))
sns.boxplot(x='pclass',y='age',data=df,palette='PuBuGn_r')
```

```python
total_counts = df['sex'].value_counts()
survived = df[df['survived']==1]['sex'].value_counts()
dead = df[df['survived']==0]['sex'].value_counts()
df = pd.DataFrame([survived,dead])
df.index = ['survived','Dead']
df.plot(title="Survival Count by Gender", kind='barh',stacked=True)

print ("Total on Board:")
print (total_counts)
print ("")
print ("Total Survived:")
print (survived)
```
