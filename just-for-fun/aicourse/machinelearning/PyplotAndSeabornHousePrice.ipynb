```python
unfinished_basement = (df['BsmtUnfSF']).sum()
print(unfinished_basement)
finished_basement_1 = (df['BsmtFinSF1']).sum()
print(finished_basement_1)
finished_basement_2 = (df['BsmtFinSF2']).sum()
print(finished_basement_2)
```

```python
labels = ['Unfinished-Basement', 'Finished-Basement-1', 'Finished-Basement-2']
sizes = [unfinished_basement, finished_basement_1, finished_basement_2]
plt.pie(sizes, labels=labels, autopct='%1.2f%%')
plt.axis('equal')
plt.title('Distribution depending on the condition of the basement')
plt.show()
```

```python
quality_above_5_sold = ((df['OverallQual'] >= 6) & (df['HousePrice'] == 1)).sum()
print(quality_above_5_sold)
quality_above_5_unsold = ((df['OverallQual'] >= 6) & (df['HousePrice'] == 0)).sum()
print(quality_above_5_unsold)
quality_equal_or_below_5_sold = ((df['OverallQual'] <= 5) & (df['HousePrice'] == 1)).sum()
print(quality_equal_or_below_5_sold)
quality_equal_or_below_5_unsold = ((df['OverallQual'] <= 5) & (df['HousePrice'] == 0)).sum()
print(quality_equal_or_below_5_unsold)

data_sold = (quality_equal_or_below_5_sold, quality_above_5_sold)
data_unsold = (quality_equal_or_below_5_unsold, quality_above_5_unsold)
p1 = plt.bar(np.arange(2), (data_sold), width=0.4)
p2 = plt.bar(np.arange(2), (data_unsold), bottom=data_sold, width=0.4)
plt.xticks(np.arange(2), ['Sold', 'Unsold'])
plt.legend((p1[0], p2[0]), ('Below', 'Above'))
plt.show()
```

```python
plt.figure(figsize=(5,5), dpi=100)

above_10 = df.loc[df['TotRmsAbvGrd'] >= 10]['OverallQual']
above_6 = df.loc[(df['TotRmsAbvGrd'] >= 6) & (df['TotRmsAbvGrd'] < 10)]['OverallQual']
below_6 = df.loc[df['TotRmsAbvGrd'] < 6]['OverallQual']

bp = plt.boxplot([above_10, above_6, below_6], labels=['above_10', 'above_6', 'below_6'], patch_artist=True)
plt.title('House-Price-TotRmsAbvGrd')
plt.ylabel('OverallQual')

for box in bp['boxes']:
    box.set(color='#6a0dad', linewidth=2)
    box.set(facecolor='#008000')

plt.show()
```
