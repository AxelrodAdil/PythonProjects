```python
col= df.columns.tolist()
col
```

```python
x= df.drop('Survived', axis=1)
y= df['Survived']
```

```python
for i in col:
    if df[i].dtype.name == 'object':
        x[i]= pd.factorize(df[i])[0]

# from sklearn.preprocessing import LabelEncoder
# le=LabelEncoder()
# df.Name=le.fit_transform(df.Name)
# df.Sex=le.fit_transform(df.Sex)
# df.Cabin=le.fit_transform(df.Cabin)
# df.Ticket=le.fit_transform(df.Ticket)
# df.Embarked=le.fit_transform(df.Embarked)
```

```python
mean_value= x['Age'].mean()
x['Age'].fillna(value= mean_value, inplace=True)
x.head(20)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.ensemble import GradientBoostingClassifier, GradientBoostingRegressor
from sklearn.ensemble import AdaBoostClassifier, AdaBoostRegressor

from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
line_regr = LinearRegression()
```

```python
log_regr = LogisticRegression(max_iter=1000)
```

```python
DTC= DecisionTreeClassifier()
DTR= DecisionTreeRegressor()
```

```python
RFC= RandomForestClassifier(n_estimators=5,
                            max_depth=3,
                            max_features='auto',
                            min_samples_leaf=4,
                            bootstrap=True,
                            n_jobs=-1,
                            random_state=0)
RFR= RandomForestRegressor(n_estimators=5)
```

```python
GBC= GradientBoostingClassifier()
GBR= GradientBoostingRegressor()
```

```python
ABC= AdaBoostClassifier()
ABR= AdaBoostRegressor()
```

```python
import xgboost as xgb
XGBC= xgb.XGBClassifier()
XGBR= xgb.XGBRegressor()
```

```python
excel_df= pd.DataFrame(columns= ['Algorithm', 'Accuracy', 'Precision', 'Recall', 'f1', 'Confusion_matrix'])
excel_df1= pd.DataFrame(columns= ['Algorithm', 'MSE', 'MAE', 'R2'])
```

```python
def AddClassifierData(algo):
    list_row = {        
        'Algorithm': algo,
        'Accuracy': metrics.accuracy_score(y_test, y_pred),
        'Precision': metrics.precision_score(y_test, y_pred),
        'Recall': metrics.recall_score(y_test, y_pred),
        'f1': metrics.f1_score(y_test, y_pred),
        'Confusion_matrix': ' '.join(map(str, metrics.confusion_matrix(y_test, y_pred).flatten()))
    }

    df1 = pd.DataFrame(list_row, index=[0])
    df2 = pd.concat([excel_df, df1], ignore_index = True)
    return df2
```

```python
def AddRegressorData(algo):
    list_row = {        
        'Algorithm': algo,
        'MSE': metrics.mean_squared_error(y_test, y_pred),
        'MAE': metrics.mean_absolute_error(y_test, y_pred),
        'R2': metrics.r2_score(y_test, y_pred),
    }

    df1 = pd.DataFrame(list_row, index=[0])
    df2 = pd.concat([excel_df1, df1], ignore_index = True)
    return df2
```

```python
DTC.fit(x_train, y_train)
y_pred = DTC.predict(x_test)
excel_df = AddClassifierData('DecisionTreeClassifier').copy()
DTR.fit(x_train, y_train)
y_pred = DTR.predict(x_test)
excel_df1 = AddRegressorData('DecisionTreeRegressor').copy()
```

```python
GBC.fit(x_train, y_train)
y_pred = GBC.predict(x_test)
excel_df = AddClassifierData('GradientBoostingClassifier').copy()
GBR.fit(x_train, y_train)
y_pred = GBR.predict(x_test)
excel_df1 = AddRegressorData('GradientBoostingRegressor').copy()
```

```python
ABC.fit(x_train, y_train)
y_pred = ABC.predict(x_test)
excel_df = AddClassifierData('AdaBoostClassifier').copy()
ABR.fit(x_train, y_train)
y_pred = ABR.predict(x_test)
excel_df1 = AddRegressorData('AdaBoostRegressor').copy()
```

```python
line_regr.fit(x_train, y_train)
y_pred = line_regr.predict(x_test)
excel_df1 = AddRegressorData('LinearRegression').copy()
log_regr.fit(x_train, y_train)
y_pred = log_regr.predict(x_test)
excel_df = AddClassifierData('LogisticRegression').copy()
```

```python
XGBC.fit(x_train, y_train)
y_pred = XGBC.predict(x_test)
excel_df = AddClassifierData('XGBClassifier').copy()
XGBR.fit(x_train, y_train)
y_pred = XGBR.predict(x_test)
excel_df1 = AddRegressorData('XGBRegressor').copy()
```

```python
RFC.fit(x_train, y_train)
y_pred = RFC.predict(x_test)
excel_df = AddClassifierData('RandomForestClassifier').copy()
RFR.fit(x_train, y_train)
y_pred = RFR.predict(x_test)
excel_df1 = AddRegressorData('RandomForestRegressor').copy()
```
